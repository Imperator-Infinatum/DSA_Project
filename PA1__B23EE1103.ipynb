```python
def sum(list):
    total = 0
    for num in list:
        if num % 2 == 0:
            total += num                                      # add all even number
    return total

a = input("Enter a List: ")
numbers = []                                                     # Initialize an empty list
for num in a.split(" "):                                         # Split the input string by spaces
 numbers.append(int(num))                                        # Convert each element to an integer and add to the list

result = sum(numbers)

print("Sum of even numbers:", result)

```

```python
import pandas as pd

def replace_missing(file):                                                      # Function to replace missing values with column mean
    df = pd.read_csv(file, header=None)                                         # Load the CSV file without border

                                                                                 # Assign default column names (since there is no header in the file)
    df.columns = ['column1', 'column2', 'column3', 'column4', 'column5']

    print("Columns in the CSV file:", df.columns)                               # Print actual column names


    for column in df.columns:                                                   # Iterate over the columns in the DataFrame
         df[column] = df[column].replace("", float("nan"))                       # Replace empty strings with NaN
         column_mean = df[column].mean()                                          # Calculate the mean of the column (ignores NaN values)
         df[column] = df[column].fillna(column_mean)                             # Replace missing values (NaN) with the column mean


    return df

file = '/content/feature.csv'

result_df = replace_missing(file)
if result_df is not None:
    print("Data after replacing missing values:")  # Print the result after processing
    print(result_df)
```

```python
def find_name(phonebook, digit):

    matching = []
    for name, number in phonebook.items():
        if str(number).startswith(digit):      # Convert the phone number to a string and check if it starts with the prefix
            matching.append(name)
    return matching



phonebook = {
    "Dipu": 9819093072,        # Create a phonebook with your friend's data
    "Rohit": 9321976213,
    "Papa": 9821672558,
    "Anurag": 9321566468,
    "Kaniskh": 9024314139,
    "Siddharth": 8100496140,
    "Aaryan": 9321527509,
    "Pragna": 9490328976,
    "Pragati": 6310944756
}

digit = input("Enter the two-digit:-")

result = find_name(phonebook, digit)


if result:
    print(f"Names with phone numbers starting with {digit}: {', '.join(result)}")
else:
    print(f"No phone numbers found starting with {digit}.")
```

```python
import numpy as np

def relu(vector):
   return np.maximum(0, vector)       #Compute the ReLU activation for each element in the vector.

def sigmoid(vector):
  return 1 / (1 + np.exp(-vector))    #Compute the sigmoid

vector_input = input("Enter the vector element: ")
vector = np.array([float(x) for x in vector_input.split()])                 #split by space vector delete

relu_result = relu(vector)
sigmoid_result = sigmoid(vector)

print("ReLU Activation: ", relu_result)
print("Sigmoid Activation: ", sigmoid_result)

```

```python
import cv2
import matplotlib.pyplot as plt

def simple_average_redness(image_path):

    image = cv2.imread(image_path)

    if image is None:
        print("Error: Unable to read the image. Check the file path.")
        return None
                                                                                 # Convert the image from BGR to RGB for proper color display in matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

                                                                                 # Display the image using matplotlib
    plt.imshow(image_rgb)
    plt.axis('off')                                                              # Turn off the axis
    plt.show()

                                                                                # Extract the red channel directly (BGR format: Blue, Green, Red)
    red_channel = image[:, :, 2]                                                # Red channel is the last in BGR

    return red_channel.mean()


image_path = "/content/apple-1834639_1280.jpg"

average_redness = simple_average_redness(image_path)
if average_redness is not None:
    print(f"Average Redness: {average_redness:.2f}")
```

```python
import os
import cv2
from google.colab.patches import cv2_imshow                                     # Use this to display images in Colab

def display(image_path, width=100, height=100):

    image = cv2.imread(image_path)
    if image is not None:
        resized_image = cv2.resize(image, (width, height))
        cv2_imshow(resized_image)
    else:
        print(f"Unable to load image: {image_path}")

def find(folder_path):

    for root, dirs, files in os.walk(folder_path):                              # Traverse the directory and subdirectories
        for file in files:
                                                                                # Check if the file is an image (by extension)
            if file.lower().endswith(('.png', '.jpeg', '.jpg', '.tiff')):
                image_path = os.path.join(root, file)
                print(f"Found image: {image_path}")
                display(image_path)

folder_path = "/content/sample_data"  # Example folder path

find(folder_path)
```

```python
import math

class Circle:
    def __init__(self, radius):
        if radius < 0:
            raise ValueError("Radius cannot be negative")
        self.radius = radius

    def calculate_area(self):
        return math.pi * (self.radius ** 2)

    def calculate_perimeter(self):
        return 2 * math.pi * self.radius

x1= Circle(5)
x2 = Circle(10)

print(f"Circle with radius {x1.radius}: Area = {x1.calculate_area():.2f}, Perimeter = {x1.calculate_perimeter():.2f}")
print(f"Circle with radius {x2.radius}: Area = {x2.calculate_area():.2f}, Perimeter = {x2.calculate_perimeter():.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "/content/marks.csv"
df = pd.read_csv(file_path)

df['Average'] = df.iloc[:, 1:].mean(axis=1)                                     # Calculate average and variance
df['Variance'] = df.iloc[:, 1:].var(axis=1)
students = df['Name']

plt.figure(figsize=(16, 8))                                                     # Increased width for better spacing
                                                                                # Bar width and positions
x = range(len(students))
width = 0.4

# Plot averages and variances side-by-side
plt.bar(x, df['Average'], width=width, label='Average', color='white', edgecolor='black')
plt.bar([p + width for p in x], df['Variance'], width=width, label='Variance', color='black', edgecolor='black')

# Add labels and title
plt.xlabel('Students', fontsize=14)
plt.ylabel('Scores', fontsize=14)
plt.title('Average and Variance of Scores', fontsize=16)

                                                                                # Rotate and space out x-axis labels for clear name visible
plt.xticks([p + width / 2 for p in x], students, rotation=90, ha='center', fontsize=10)


plt.grid(axis='y', linestyle='--', alpha=0.7)                                   # Add gridlines
                                                                                # Add legend Displays the legend to distinguish between average and variance bars.
plt.legend()
                                                                                # Adjust layout to prevent clipping
plt.tight_layout()
plt.show()
```

```python
def count(file_path):
    with open(file_path, 'r') as file:                                           # Open the file and read its content
        lines = file.readlines()                                                 # Read all lines from the file

    word_count = {}                                                              # Dictionary to store word frequencies

    for line in lines:
        words = line.split()                                                     # Split the line into words
        for word in words:
            word = word.lower().strip(".,!?\"'")                                 # Normalize words
            word_count[word] = word_count.get(word, 0) + 1                       # Count word frequency

    for word, count in word_count.items():
        print(f"{word}: {count}")

file_path = "/content/sample.txt"                                                        # Replace with your text file name
count(file_path)
```
